# Step 2: Data Cleaning

```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
# Removing redundant columns that exist in the players table
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
# Duplicate columns removed
mvps.head()
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

Our objective is to 'clean' the players table so we can then merge it with the mvps table by player and year.

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
# In order to merge them, we need to remove the asterisk at the end of some of the names
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head()
```

Some players like Greg Anderson played for different teams in the same season, for eg. in 1991. We want it so that each player gets one row in the merged table.


```python
players[players["Player"] == "Greg Anderson"]
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy()
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
sorted(teams["Team"].unique())
```

```python
sorted(combined["Tm"].unique())
```

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
train = combined.merge(teams, how="outer",on=["Team", "Year"])
```

```python
train
```

```python
del train["Unnamed: 0"]
```

```python
train.dtypes
```

```python
train = train.apply(pd.to_numeric, errors='ignore')
```

```python
train.dtypes
```

```python
train["GB"].unique()
```

```python
train["GB"] = pd.to_numeric(train["GB"].str.replace("—", "0"))
```

```python
train.dtypes
```

```python
# Exporting the final table that is ready for machine learning to a csv file
train.to_csv("player_mvp_stats.csv")
```
