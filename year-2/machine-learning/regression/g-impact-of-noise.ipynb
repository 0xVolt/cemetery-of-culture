```python
# Import libraries
from sklearn.datasets import make_regression
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create noise and score vectors
noiseVector = np.arange(0, 110, 10)
scoreVector = []
```

```python
# Use sklearn to calculate and store the r2 scores 
for i in noiseVector:
    # Make a new regression model
    X, T = make_regression(1000, 5, n_targets=2, shuffle=True, bias=0, noise=i, random_state=5)
    reg = LinearRegression().fit(X, T)
    
    # Calculate the y_pred vector
    yPredict = reg.predict(X)
    
    # Append our score to the score vector
    scoreVector.append(r2_score(T, yPredict))
```

```python
# Labelling the axes
plt.xlabel('Noise')
plt.ylabel('R2 Score')

# Generate and display plot
plt.plot(noiseVector, scoreVector)
plt.show()
```
