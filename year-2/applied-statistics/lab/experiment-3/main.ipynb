## Working with Normal Gaussian Distributions

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a function to calculate the respective Z-Score for every data point
def calcZScore(x, mean, std):
    return ((x - mean) / std)
```

```python
numberSample = 1000

# Generate the array using the numpy.random.normal(mean, std, numberOfElts) method
sampleArray = np.random.normal(5, 2, numberSample)
```

```python
# Calculate mean and std. deviation of the generated array
sampleArrayMean = np.mean(sampleArray)
sampleArrayStd = np.std(sampleArray)
sampleArrayVar = np.var(sampleArray)

np.sort(sampleArray)

# Give the summary statistics
print(
    "\nGenerated Sample List with {} elements:\n\n{}\n\n------SUMMARY------\n\nCalculated Mean = {:.3f}\nCalculated Standard Deviation = {:.3f}\nCalculated Variance = {:.3f}\n".format(
        numberSample, sampleArray, sampleArrayMean, sampleArrayStd, sampleArrayVar
    )   
)
```

```python
# Plotting the graph
# Setting a grid
plt.grid()

# Setting graph title
plt.title("Normal distribution w/ Mean: 5 & Std. Deviation: 2")

# Setting axes labels
plt.xlabel('x')
plt.ylabel('Normal Distribution')

# Plotting and displaying graph
plt.hist(sampleArray)
plt.show()
```

```python
# Convert the above generated normally distributed data to Z-distributed data using list comprehension
zDistributedData = [calcZScore(i, sampleArrayMean, sampleArrayStd) for i in sampleArray]

# Plot two graphs
plt.title("Normal distribution w/ Mean: 5 & Std. Deviation: 2")
plt.xlabel('x')
plt.ylabel('Normal Distribution')
plt.grid()
plt.hist(sampleArray)
plt.show()

plt.title("Z - Distribution for above graph")
plt.xlabel('x')
plt.ylabel('Standard Normal / Z - Distribution')
plt.grid()
plt.hist(zDistributedData)
plt.show()
```

```python
retList = checkSkewness(zDistributedData)
print('The above normal gaussian distribution is', retList[0], 'skewed')
```
