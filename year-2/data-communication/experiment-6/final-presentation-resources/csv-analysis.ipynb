```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data generated as a .csv file into pandas dataframe
data = pd.read_csv('out.csv')

# Create a dataframe with the imported data
df = pd.DataFrame(data, columns= ['Frame (#)', 'Attempts (k)', 'Total Time Taken (seconds)', 'Backoff Time (seconds)'])

# Set the first column to be the index
df.set_index('Frame (#)')
```

```python
# Describing the dataframe that imports our csv data
print(df.describe())
```

### Analysis #1 - Attempts (k) vs. Frames

```python
# Plotting Frames vs. Attempts
frames = df['Frame (#)']
attempts = df['Attempts (k)']

# Setting up labels for axes
plt.title('Attempts (k) vs. Frame')
plt.xlabel('Frame')
plt.ylabel('Attempts (k)')

# Plotting graph
plt.plot(frames, attempts, color="lightslategray")
plt.show()

# Printing summary statistics
print("Mean number of attempts: ", np.mean(attempts))
```

### Analysis #2 - Distribution of k

```python
plt.title("Distribution of Attempts (k)")

plt.xlabel("k")
plt.ylabel("Frequency of occurrence")

plt.hist(attempts, edgecolor="white", color="lightslategray")
plt.show()
```

### Analysis #3 - Total Time Taken vs. Frames

```python
totalTime = df['Total Time Taken (seconds)']

plt.title("Total Time Taken (s) vs. Frames")
plt.xlabel("Frames")
plt.ylabel("Total Time Taken (s)")

plt.plot(frames, totalTime, color ='lightsteelblue')
plt.show()
```

### Analysis #4 - Distribution of Total Time Taken

```python
plt.title("Distribution of Total Time Taken (s)")

plt.xlabel("Time (s)")
plt.ylabel("Frequency of occurrence")

plt.hist(totalTime, edgecolor="white", color ='lightsteelblue')
plt.show()
```

### Analysis #5 - Attempts vs. Total Time Taken

```python
plt.title("Attempts vs. Total Time Taken")

plt.xlabel("Total Time Taken")
plt.ylabel("Attempts (k)")

plt.plot(totalTime, attempts, color ='orange')
plt.show()
```
