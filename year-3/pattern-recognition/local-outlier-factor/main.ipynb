```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Load the dataset
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
X = iris.iloc[:, :-1].values

# Define LOF model
model = LocalOutlierFactor(n_neighbors=20, contamination='auto')

# Fit the model and get outlier scores
y_pred = model.fit_predict(X)
lof_scores = -model.negative_outlier_factor_

# Visualize outlier scores
plt.figure(figsize=(10, 5))
plt.scatter(range(len(lof_scores)), lof_scores, c=y_pred, cmap='coolwarm')
plt.xlabel('Sample Index')
plt.ylabel('LOF Score')
plt.title('Local Outlier Factor')
plt.show()
```
