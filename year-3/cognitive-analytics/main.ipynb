```python
import matplotlib.pyplot as plt
import numpy as np

# Set up figure and subplots
# fig, axs = plt.subplots(3, 3)

# Scatter plot
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y)

```

```python

# Line plot
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python

# Bar plot
x = ["A", "B", "C", "D"]
y = [1, 2, 3, 4]
plt.bar(x, y)
```

```python

# Histogram
x = np.random.normal(0, 1, 100)
plt.hist(x)
```

```python

# Pie chart
labels = ["A", "B", "C", "D"]
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")

```

```python

# Box plot
x = np.random.normal(0, 1, 100)
plt.boxplot(x)
```

```python

# Violin plot
x = np.random.normal(0, 1, 100)
plt.violinplot(x)

```

```python
# Contour plot
def f(x, y):
    return np.sin(x) + np.cos(y)

x = np.linspace(0, 2 * np.pi, 100)
y = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z)

```

```python
# 3D scatter plot
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(0, 1, 100)
plt.scatter3D(x, y, z)
```
