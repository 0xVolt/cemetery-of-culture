# Experiment 2 - Gradient Descent

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Gradient Descent on a quadratic function

### Function to implement the gradient descent algorithm

```python
def gradient_descent(start, gradient, learning_rate, max_iterations, tolerance=0.01):
    # To track the history of the algorithm
    trace = [start]
    optimiserValue = start
    
    # To count the number of iterations
    iterations = 0

    for _ in range(max_iterations):
        diff = learning_rate * gradient(optimiserValue)
        
        if np.abs(diff) < tolerance:
            break
        
        optimiserValue = optimiserValue - diff
        
        # Track the value and append it to the list
        trace.append(optimiserValue)
        
        iterations += 1
        
    return trace, optimiserValue, iterations
```

### Defining our test function and it's gradient

```python
def function(x):
    return x ** 2 - 4 * x + 1

def gradient(x):
    return 2 * x - 4
```

### Calling the gradient descent function

Defining the hyper-parameters for the gradient descent function

```python
start = 100
learning_rate = 0.001
max_iterations = 1000
```

Calling the function

```python
history, result, iterations = gradient_descent(start=start, gradient=gradient, learning_rate=learning_rate, max_iterations=max_iterations)
```

### Tracing the change in the gradient value over iterations.

```python
print(f'Tracing the gradients:\n{history}\n\nFinal result of the gradient descent algorithm: {result}')
```

### Plotting the change in gradient

```python
plt.title('Change in Gradient Value')

plt.xlabel('Iteration')
plt.ylabel('Gradient Value')

plt.grid()

plt.plot(history)
```

### Plotting the descent over the function

Defining the axes and the scale of the x-axis

```python
xScale = 1000

X = np.arange(-xScale, xScale + 1)
y = [function(i) for i in X]
```

```python
plt.grid()

plt.plot(history)
plt.plot(X, y)
```
